```python
### multiple regression analysis

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Assuming my 'df'  DataFrame containing the data

# Extracting features 'RM' and 'LSTAT', and target variable 'MEDV'
x = df[['RM', 'LSTAT']].values
y = df['MEDV'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)

# Create a Linear Regression model
LR = LinearRegression()

# Fit the model on the training data
LR.fit(X_train, y_train)

# Make predictions on the training and test sets
y_train_pred = LR.predict(X_train)
y_test_pred = LR.predict(X_test)

# Evaluate the model using R-squared (coefficient of determination)
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

# Printing the R-squared scores for the training and test sets
print('R2_train= ', train_r2, '\nR2_test= ', test_r2)
```

```python
### multiple regression analysis 
    ### All cols.
# Importing the necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Assuming  the 'df'  DataFrame containing the data

# Extracting features (excluding 'MEDV') and target variable 'MEDV'
x = df.drop("MEDV", axis=1).values
y = df['MEDV'].values

# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)

# Creating a Linear Regression model
LR = LinearRegression()

# Fit the model on the training data
LR.fit(X_train, y_train)

# Make predictions on the training and test sets
y_train_pred = LR.predict(X_train)
y_test_pred = LR.predict(X_test)

# Evaluate the model using R-squared
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

# Printing the R-squared scores for the training and test sets
print('R2_train= ', train_r2, '\nR2_test= ', test_r2)
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming my 'df' DataFrame
x = df[['CRIM', 'ZN', 'INDUS', 'CHARS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns  # Set the "feature" column correctly

# Calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]  # Calculate VIF properly

print(vif_data)
```
